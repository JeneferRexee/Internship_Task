```python
import os
import fitz
import pdfplumber
import re

def extract_pages(pdfpath, lib='plumber'):
    if not os.path.exists(pdfpath):
        print("PDF path does not exist.")
        return None

    all_txt = []

    # FITZ extraction 
    if lib == 'fitz':
        try:
            doc = fitz.open(pdfpath)
            print(f'Number of pages (fitz): {len(doc)}')

            for i, page in enumerate(doc):
                txt = page.get_text()
                if txt:
                    clean_txt = re.sub(r"[^A-Za-z0-9 ]", "", txt)
                    all_txt.append(clean_txt)
                else:
                    print(f"Page {i+1} may be a scanned image or empty.")
                    all_txt.append("")
            doc.close()
            return all_txt

        except Exception as e:
            print("FITZ ERROR:", e)
            return None

    # PDFPLUMBER extraction
    elif lib == 'plumber':
        try:
            with pdfplumber.open(pdfpath) as pdf:
                print(f'Number of pages (plumber): {len(pdf.pages)}')

                for i, page in enumerate(pdf.pages):
                    txt = page.extract_text()
                    if txt:
                        clean_txt = re.sub(r"[^A-Za-z0-9 ]", "", txt)
                        all_txt.append(clean_txt)
                    else:
                        print(f"Page {i+1} may be a scanned image or empty.")
                        all_txt.append("")
            return all_txt

        except Exception as e:
            print("PLUMBER ERROR:", e)
            return None

    else:
        print("Invalid library name. Use 'fitz' or 'plumber'.")
        return None
```

```python
pdfpath = r"C:\Users\Jenefer.RexeeGeorge\Downloads\AI_11_ISC_2 1.pdf"

texts = extract_pages(pdfpath, lib='plumber')

for i, page_text in enumerate(texts):
    print(f"Page {i+1}:\n{page_text}\n{'-'*50}")
```
